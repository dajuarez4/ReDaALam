```python
from utils.config import simulations_path, potentials_path, potentials, experiments_path, executables_path, executable, atom_type1, atom_type2
from utils.default_dictionaries import base_params_atoms_position, defect_percentages, reduction_percentages,gen_lammps_input  
import os
import shutil
import subprocess
```

```python
print(simulations_path)
reduction_percentages = [0.01]
simulations_path ='/global/cfs/projectdirs/m3845/Diego_projects/NiTi/Phonons/molecular_dynamics_1/'
```

```python

n_temp_variations = 1
initial_temperature = 10
lattice_parameter = 3.00
PTE_path =  '/global/cfs/projectdirs/m3845/Diego_projects/NiTi/Phonons/new_test/UTEP-tedehamo/PTE/'
PTE_files = ['save-data.py','plot-1.py']
images_path = '/global/cfs/projectdirs/m3845/Diego_projects/NiTi/Phonons/new_test/UTEP-tedehamo/Figures/'
for conc in reduction_percentages: # per concentration reduction
    concentration_path_sim = simulations_path + f'{(0.50+conc):.2f}-{atom_type1}-{(0.50-conc):.2f}-{atom_type2}/'
    os.makedirs(concentration_path_sim, exist_ok=True)
    for defect in defect_percentages: # for vacancy change to vacancy_percentages
        defect_path_sim = concentration_path_sim + f'defect_{defect}%'
        for temperature_variation in range(n_temp_variations):
                temperature = temperature_variation * 100 + initial_temperature # ---------- temp incresead by 100 degrees across ntemp iterations
                directory = defect_path_sim + f'/{atom_type1}{atom_type2}_T_{int(temperature)}_L_{lattice_parameter:.2f}'
                os.makedirs(directory, exist_ok=True)
                for PTE in PTE_files:
                    src = os.path.join(PTE_path, PTE)
                    dest = os.path.join(directory, PTE)
                    shutil.copy(src, dest)
                    subprocess.run(['python', f'{PTE}'], cwd=directory)
                
                src2 = os.path.join(directory, "all_plots_2x2.png")
                dst2 = os.path.join(images_path, f"Plot-{(0.50+conc):.2f}-{atom_type1}-{(0.50-conc):.2f}-{atom_type2}-T-{int(temperature)}.png") 
                shutil.copy(src2, dst2)
                                        
```
